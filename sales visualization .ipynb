```python
df = pd.read_csv("Car_sales.csv")
```

```python
df["__year_resale_value"].fillna(df["__year_resale_value"].median(), inplace=True)
df["Price_in_thousands"].fillna(df["Price_in_thousands"].median(), inplace=True)
df["Engine_size"].fillna(df["Engine_size"].median(), inplace=True)
df["Horsepower"].fillna(df["Horsepower"].median(), inplace=True)
df["Wheelbase"].fillna(df["Wheelbase"].median(), inplace=True)
df["Width"].fillna(df["Width"].median(), inplace=True)
df["Length"].fillna(df["Length"].median(), inplace=True)
df["Curb_weight"].fillna(df["Curb_weight"].median(), inplace=True)
df["Fuel_capacity"].fillna(df["Fuel_capacity"].median(), inplace=True)
df["Fuel_efficiency"].fillna(df["Fuel_efficiency"].median(), inplace=True)
df["Power_perf_factor"].fillna(df["Power_perf_factor"].median(), inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df["Price_in_thousands"])
plt.title("Price in Thousands", size=15)
plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
lowerprice, upperprice = remove_outlier(df["Price_in_thousands"])
df["Price_in_thousands"] = np.where(df["Price_in_thousands"]>upperprice, upperprice, df["Price_in_thousands"])
df["Price_in_thousands"] = np.where(df["Price_in_thousands"]<lowerprice, lowerprice, df["Price_in_thousands"])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(df["Price_in_thousands"])
plt.title("Price in Thousands Without Outliers", size=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Horsepower", y="Sales_in_thousands", data=df)
plt.title("Correlation between Horse Power and Sales", size=18)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Engine_size", y="Sales_in_thousands", data=df)
plt.title("Correlation between Engine Size and Sales", size=18)
plt.show()
```

```python
price = df["Price_in_thousands"]*1000
plt.figure(figsize=(10,6))
sns.distplot(price, color="g")
plt.title("Density of Car Prices", size=15)
plt.xlabel("Price", size=13)
plt.ylabel("Density",size=13)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True, square=True, cmap="RdYlBu")
plt.title("Correlation Between Variables", size=18)
plt.xticks(size=10)
plt.yticks(size=10)
plt.show()
```
